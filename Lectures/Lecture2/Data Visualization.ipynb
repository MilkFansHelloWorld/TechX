# Data Visualization Demo

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Line plot**

```python
x = [x*0.1 for x in range(100)]
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

**Scattered plot**

```python
plt.scatter(x,y)
plt.show()
```

**Bar Chart**

```python
u=['Red','Blue','Green']
v=[1,4,2]
```

```python
plt.bar(u,v)
plt.show()
```

Change colors

```python
plt.bar(u,v,color=('r','b','g'))
plt.show()
```

**How about adding having two graphs in the figures?**

```python
plt.figure(1)
plt.plot(x,y)
plt.figure(2)
plt.bar(u,v,color=('r','b','g'))
plt.show()
```

**You can even have them appear on the same graph**

```python
plt.plot(x,y)
plt.plot(x,np.cos(x),'r--')
plt.show()
```

**Add titles and legends!**

```python
plt.figure(1)
plt.plot(x,y)
plt.plot(x,np.cos(x),'r--')
plt.title('this is the title of the graph')
plt.xlabel('x-label')
plt.ylabel('cos/sin')
plt.show()
```
